# Tutorial on MEEP simulation run with phcr-cavity.py
The tutorial is based on __[documentation example](https://meep.readthedocs.io/en/latest/Python_Tutorials/Resonant_Modes_and_Transmission_in_a_Waveguide_Cavity/)__. 

To run the simulation we have to address the python translator by running system command with <b>!</b> symbol. 

We run simulation code <b>phcr-cavity.py</b>

The console output is saved at <b>Output.txt</b>

````python
from subprocess import call
import numpy as np
import matplotlib.pyplot as plt
import math as mt

# simfileName ='freqs_1.dat'
# fileNameNorm_disp = 'dispersion_relations_1.dat'
# fileNameNorm_w = 'Omega.dat'

# ex_str_d =f'grep zevenyoddfreqs: {simfileName}  | cut -d , -f3,7-| sed 1d> {fileNameNorm_disp}'
# ex_str_w =f'grep zevenyoddfreqs: {simfileName}  | cut -d , -f7-| sed 1d> {fileNameNorm_w}'

# call(ex_str_d, shell="True")
# call(ex_str_w, shell="True")

a = 0.43         # units of um
r = 0.12         # units of um
h = 0.22         # units of um
w = 0.50         # units of um

r = r/a          # units of "a"
h = h/a          # units of "a"
w = w/a          # units of "a"
c = 3e8         # units of um/s 

am = 0.43e-6
g = 2*np.pi/am


# ------------------------------------------------------------------------------------------------------------------------------------


fig, ax = plt.subplots()
f = np.genfromtxt("modes_1.dat", delimiter=",")
ax.plot(f[:,0],f[:,1:],'b-', linewidth = 3)
ax.plot(f[:,0],f[:,0],'g-', linewidth = 3, linestyle = '--', label='Light line, $sc_y$ = 4, $sc_z$ = 4')
plt.xlabel("Wavevector, $k_x a/2\pi$, a.u.", size=16)
plt.ylabel("Frequency, $\omega a/2\pi c$, a.u.", size=16)
# plt.axis([0,0.5,0,0.4])
# plt.xticks([t for t in np.arange(0,0.6,0.1)])
# plt.yticks([t for t in np.arange(0,0.5,0.1)])
ax.grid(True)
plt.legend()
plt.show()

fig, ax = plt.subplots()
f = np.genfromtxt("modes_2.dat", delimiter=",")
ax.plot(f[:,0],f[:,1:],'b-', linewidth = 3)
ax.plot(f[:,0],f[:,0],'g-', linewidth = 3, linestyle = '--', label='Light line, $sc_y$ = 0.5, $sc_z$ = 0.22')
plt.xlabel("Wavevector, $k_x a/2\pi$, a.u.", size=16)
plt.ylabel("Frequency, $\omega a/2\pi c$, a.u.", size=16)
# plt.axis([0,0.5,0,0.4])
# plt.xticks([t for t in np.arange(0,0.6,0.1)])
# plt.yticks([t for t in np.arange(0,0.5,0.1)])
ax.grid(True)
plt.legend()
plt.show()

fig, ax = plt.subplots()
f = np.genfromtxt("modes_3.dat", delimiter=",")
ax.plot(f[:,0],f[:,1:],'b-', linewidth = 3)
ax.plot(f[:,0],f[:,0],'g-', linewidth = 3, linestyle = '--', label='Light line, $sc_y$ = 1, $sc_z$ = 1')
plt.xlabel("Wavevector, $k_x a/2\pi$, a.u.", size=16)
plt.ylabel("Frequency, $\omega a/2\pi c$, a.u.", size=16)
# plt.axis([0,0.5,0,0.4])
# plt.xticks([t for t in np.arange(0,0.6,0.1)])
# plt.yticks([t for t in np.arange(0,0.5,0.1)])
ax.grid(True)
plt.legend()
plt.show()

fig, ax = plt.subplots()
f = np.genfromtxt("modes_4.dat", delimiter=",")
ax.plot(f[:,0],f[:,1:],'b-', linewidth = 3)
ax.plot(f[:,0],f[:,0],'g-', linewidth = 3, linestyle = '--', label='Light line, $sc_y$ = 2, $sc_z$ = 2')
plt.xlabel("Wavevector, $k_x a/2\pi$, a.u.", size=16)
plt.ylabel("Frequency, $\omega a/2\pi c$, a.u.", size=16)
# plt.axis([0,0.5,0,0.4])
# plt.xticks([t for t in np.arange(0,0.6,0.1)])
# plt.yticks([t for t in np.arange(0,0.5,0.1)])
ax.grid(True)
plt.legend()
plt.show()

fig, ax = plt.subplots()
f = np.genfromtxt("modes_5.dat", delimiter=",")
ax.plot(f[:,0],f[:,1:],'b-', linewidth = 3)
ax.plot(f[:,0],f[:,0],'g-', linewidth = 3, linestyle = '--', label='Light line, $sc_y$ = 6, $sc_z$ = 6')
plt.xlabel("Wavevector, $k_x a/2\pi$, a.u.", size=16)
plt.ylabel("Frequency, $\omega a/2\pi c$, a.u.", size=16)
# plt.axis([0,0.5,0,0.4])
# plt.xticks([t for t in np.arange(0,0.6,0.1)])
# plt.yticks([t for t in np.arange(0,0.5,0.1)])
ax.grid(True)
plt.legend()
plt.show()

fig, ax = plt.subplots()
f = np.genfromtxt("modes_6.dat", delimiter=",")
ax.plot(f[:,0],f[:,1:],'b-', linewidth = 3)
ax.plot(f[:,0],f[:,0],'g-', linewidth = 3, linestyle = '--', label='Light line, $sc_y$ = 8, $sc_z$ = 8')
plt.xlabel("Wavevector, $k_x a/2\pi$, a.u.", size=16)
plt.ylabel("Frequency, $\omega a/2\pi c$, a.u.", size=16)
# plt.axis([0,0.5,0,0.4])
# plt.xticks([t for t in np.arange(0,0.6,0.1)])
# plt.yticks([t for t in np.arange(0,0.5,0.1)])
ax.grid(True)
plt.legend()
plt.show()

num_k = 20

v_1THz = f[num_k+1,1]*g*c/(2*np.pi*1e12)
v_2THz = f[num_k+1,2]*g*c/(2*np.pi*1e12)

lambda_1 = c/(v_1THz*1e12)
lambda_2 = c/(v_2THz*1e12)

v_1 = f[num_k+1,1]
v_2 = f[num_k+1,2]
    
print("v1=", v_1, "v2=", v_2, "v1THz=", v_1THz, "v2THz=", v_2THz)
print("lambda1=", lambda_1, "lambda2=", lambda_2)
````
